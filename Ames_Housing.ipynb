#****1.1 Background****


A key measure of ample standard of living is the availability of suitable housing. The ultimate goal towards owning a home is one of the most important decisions made by a prospective house owner which most times requires a knowledge of the cost(Demary,2009). However, the  cost of one  property in comparison to another is not often a straightforward process.

The excercise of property valuation is a nuanced process with implications for property owners, prospective buyers,and the entire real estate market. At its foundation, the process is crucial to the determination of a fair market value for a property. A fair market valuation is important for a myraid of reasons. It is a key aspect in making a decision on whether to purchase a property or not. It is also a key ingredient in property tax determination as well as mortgage credit financing.

Property valuation, is often an integrated and comprehensive process that draws upon a number of features or variables. On the one hand, it may consider exogenous features, like location, proximity to amenities, transportation links and networks, schools, social scenes (Chang and Lin, 2012). These features are the so-called 'Geographic features' (Gao et al. 2019). On the other hand, a property valuation would also consider vital property characteristics like the age of the property, the quality of finishing, available ground space, number of rooms, energy efficiency of the bilding, available facilities e.t.c These are the non-geographic features of a property (Gao et al. 2019). These features taken-together are the key ingredients industry professionals assess to arrive at a fair valuation for a property.

A popular method industry professionals may use in assessing the value of a property is the Comparative Market Analysis (CMA). In its simplest form, it determines the value of a property by benchmarking its value against the sales value of similar properties in the vicinity that may have recently sold,or are listed for sales on the real estate market. Clearly, the success of such a method would hinge on the availability of robust and readily available data on the sales value and property features.

This study is intended as an inception component of a CMA. It employs the use of python to carry out an Exploratiory Data Analysis (EDA) of The Residential Housing Prices in a USA town using data from 2006-2010.
