

##**Predicting student's dropout and academic success.**


**Abstract**

A major indicator of likely hitches in the higher education system is attributed to large number of student dropouts in the early academic years. A careful analysis of the existing data set which holds a number of demographic and socioeconomic factors regarding different students and their eventual successful completion or not of programmes in a given higher institution. To better understand obstacles responsible for these dropouts, the data are processed by the application of machine learning methods such as decision trees and random forest. The results indicate that machine learning can predict with reasonable accuracy the likelihood of success for a student (i.e graduate) however, the models suffer from an inbalance in the data class of the dataset.

#**1.0 INTRODUCTION**

Discontinuing school by students has globally become a significant challenge faced by most universities.  (Unesco,2020), reported that the interruption experienced in the educational system during COVID-19 left about 24 million students at the risk of not being able to continue their studies with higher education expected to experience the highest dropout rate and an estimated 3.5% decline rate in enrolment, which is likely to result in 7.9 million fewer students. A lot of students suffer from various struggles in school life, which ultimately cause them to give up studying (Song et al.,2023).

However, the struggles dropout students face is usually not resolved just by dropping out. Since they enter society and difficulties faced due to drop out could result to social pressure because of early entry into society and academic collapse. Unfortunately, it could result to difficulties getting employment, low income, and crime (Council of Economic Advisers,2016).

The grounds for dropping out have been associated to a number of demographic, socioeconomic and environmental reasons like school, family, society, and psychology (Contreras et al.,2022, Del Savio et al.,2022). It is agreed that an understanding of the interactions of these factors is neccessary for the improvement of students success in school.

The growth in the use of artificial intelligence has meant that tasks as such have now been undertaken widely and are well documented in the literature Therefore, this study tends to Predict student's dropout and academic success.

#**3.0 TASK DEFINITION**

The dataset contains institutional records of students with respect to their status after 'normal time' for completion. The problem is formulated as a classification task to predict students' outcome using a range of demographic and socio-economic factors known at the point of enrollment. The target class include 'Graduate' which are students who are able to complete their course in normal time, 'enrolled' which are students who after normal time are still enrolled and are likely to complete, with 'dropout' meaning students who have abandoned their program. This corresponds to three levels of risk with drop out representing the most pressing class of students who could benefit from early intervention to keep them in school and ensure their success.
The task is approached as a classification task and uses two algorithms decision tree and random forest to comparatively assess performace. Spefics about the data encoding and the features class can be found in section 5

Next, the collected demographic, socio-economic factors and student's academic path are used as features to predict performacance.

```python
dropped_columns=['Curricular units 1st sem (credited)',
       'Curricular units 1st sem (enrolled)',
       'Curricular units 1st sem (evaluations)',
       'Curricular units 1st sem (approved)',
       'Curricular units 1st sem (grade)',
       'Curricular units 1st sem (without evaluations)',
       'Curricular units 2nd sem (credited)',
       'Curricular units 2nd sem (enrolled)',
       'Curricular units 2nd sem (evaluations)',
       'Curricular units 2nd sem (approved)',
       'Curricular units 2nd sem (grade)',
       'Curricular units 2nd sem (without evaluations)']
clean_data=clean_data.drop(columns=dropped_columns)
```

```python
#Select features for dependent variable class
dependent_features=['Marital status', 'Application mode', 'Application order', 'Course',
       'Daytime/evening attendance\t', 'Previous qualification',
       'Previous qualification (grade)', 'Nationality',
       "Mother's qualification", "Father's qualification",
       "Mother's occupation", "Father's occupation", 'Admission grade',
       'Displaced', 'Educational special needs', 'Debtor',
       'Tuition fees up to date', 'Gender', 'Scholarship holder',
       'Age at enrollment', 'International',
       'Unemployment rate',
       'Inflation rate', 'GDP']
```

```python
X=clean_data[dependent_features].copy()
```

##***5.4 Perform Test and Train Split***

A proper data split is crucial to ensure that the test dataset accurately represents the variance present in the entire dataset (Jason Brownlee, 2020). A 80/20 split is adopted for this model. This split randomly partitions the data into 80% for training the model and 20% for testing, allowing the test set to have a diverse range of patterns and variations present in the entire dataset.

```python
# split data in train and test set
# X = input data, y = output labels
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
```

##***5.5 Decision Tree Classifier***

```python
#Fit the model on the training set
Success_classifier=DecisionTreeClassifier(max_leaf_nodes=10, min_samples_leaf=5, random_state=0)
Success_classifier.fit(X_train, y_train)
```

```python
#Predict with model on the test set
predictions=Success_classifier.predict(X_test)
```

```python
predictions[:10]
```

```python
y_test['Target_encoded'][:10]
```

```python
#Plotting the decision tree
#Increase the figure size and resolution
plt.figure(figsize=(16, 12), dpi=400)
plot_tree(Success_classifier, feature_names=dependent_features, filled=True, rounded=True);
```

##***5.6 Evaluation Metrics***

For classification tasks like the one in this report, metrics of precision, accuracy, recall and F1 are used widely in the literature for evaluating performance. Each metric typically complements the other. For instance Figures 1 and 2 would suggest there is an imbalance in the dataset which would mean that the measure of accuracy alone is not reliable in evaluating performance, as the algorithm may perform  extremely well in the overrepresented data in the dataset while it would perform poorly on the underrepresented data in the dataset. In such a case, the recall would highlight how many instances of performance were true positives. And further the F1-score would represent  the trade-off between recall and precision. These metrics taken together give a detailed picture of performance.

```python
#calculate accuracy score of Decision Tree
accuracy_score(y_true = y_test, y_pred = predictions)
```

```python
#Generate decision tree classification report
report = classification_report(y_test, predictions)
print('Decision Tree Classification Report:')
print(report)
```

```python
#Create a confusion matrix for the decision tree
conf_matrix = confusion_matrix(y_true=y_test, y_pred=predictions)
sns.heatmap(conf_matrix, square=True, annot=True, fmt='d', cbar=False, cmap="Greens")
plt.title('Figure 3: Confusion Matrix (Decision tree)')
plt.xlabel('Predicted value')
plt.ylabel('Actual value');
```

Figure 3 shows the confusion matrix based on the performance of the decision tree algorithm


```python
# A cross-validation of the decision tree model results is performed
#The 'cv' parameter specifies the level of folds in cross-validation.
# The 'scoring' parameter selects the metric for evaluation which in this case is accuracy.


scores = cross_val_score(Success_classifier, X, y, cv=5, scoring='accuracy')
# show the scores of the cross-validation
print("Cross-Validation Scores Decision Tree:", scores)
# shows standard deviation and the mean of the scores
print(f"Mean Accuracy: {scores.mean():.2f}")
print(f"Standard Deviation: {scores.std():.2f}")
```

##***5.7 Random Forest Classifier***

```python
# Convert y_train to a 1-dimensional array using values and ravel()
y_train_flat = y_train.values.ravel()
# Create and train a Random Forest classifier
Success_classifier_rf = RandomForestClassifier(n_estimators=100, random_state=0)
Success_classifier_rf.fit(X_train, y_train_flat)
# Make predictions on the test set
predictions_rf = Success_classifier_rf.predict(X_test)
# Calculate accuracy
accuracy_rf = accuracy_score(y_true=y_test, y_pred=predictions_rf)
print(f'Random Forest Accuracy: {accuracy_rf}')
# Create a confusion matrix
conf_matrix_rf = confusion_matrix(y_true=y_test, y_pred=predictions_rf)
print('Random Forest Confusion Matrix:')
print(conf_matrix_rf)
```

```python
#calculate accuracy score of Random Forest
accuracy_score(y_true = y_test, y_pred = predictions_rf)
```

```python
#Create a confusion matrix for the Random Forest
conf_matrix = confusion_matrix(y_true=y_test, y_pred=predictions_rf)
sns.heatmap(conf_matrix, square=True, annot=True, fmt='d', cbar=False, cmap="Greens")
plt.title('Figure 4: Confusion Matrix (Random Forest)')
plt.xlabel('Predicted value')
plt.ylabel('Actual value');
```

Figure 3 shows the confusion matrix based on the performance of the decision tree algorithm

```python
class_report_rf = classification_report(y_true=y_test, y_pred=predictions_rf)
print('Random Forest Classification Report:')
print(class_report_rf)
```

```python
# A cross-validation of the decision tree model results is performed
#The 'cv' parameter specifies the level of folds in cross-validation.
# The 'scoring' parameter selects the metric for evaluation which in this case is accuracy.


scores = cross_val_score(Success_classifier_rf, X, y, cv=5, scoring='accuracy')
# show the scores of the cross-validation
print("Cross-Validation Scores (Random Forest):", scores)
# shows standard deviation and the mean of the scores
print(f"Mean Accuracy: {scores.mean():.2f}")
print(f"Standard Deviation: {scores.std():.2f}")
```

##***6.1 Discussion on Performance***

The accuracy score for the decision tree and the random forest model were 0.62 and 0.65 respectively. This is consistent with results of the cross validation which returned a mean accuracy of 0.62 and 0.64 for decision tree and random forest respectively. This would suggest a better accuracy using the random forest. However, on close  observation of the F1- scores for the 3 classes (Dropout, Enrolled, Graduate) It is obvious that both algorithms underperform with the classification task of class 1 (enrolled). this appears consistent with Tolledo et al. (2021). This has been noted to be due to an inbalance in the representation of the dataset. While performance in the two other classes may be acceptable classification of the enrolled target would suggest that both models could benefit from further advanced resampling techniquues which is consistent with practices in the literature for where there is an inbalance in data class representation (Tanha et al., 2020)
